1. Linear Regression
Description: Models the relationship between a single dependent variable and a single independent variable using a straight line.
Use Case: When you need to predict a continuous outcome based on one predictor.
Relationship: Uncovers a linear relationship between the dependent and independent variables.

2. Multiple Linear Regression
Description: Extends linear regression to include multiple independent variables.
Use Case: When predicting a continuous outcome based on multiple predictors.
Relationship: Uncovers linear relationships between the dependent variable and multiple independent variables.

3. Logistic Regression
Description: Used for binary classification problems; models the probability of a binary outcome.
Use Case: When predicting a binary outcome (e.g., yes/no, true/false).
Relationship: Uncovers a logistic (S-shaped) relationship between the log-odds of the dependent variable and independent variables.

4. Multinomial Logistic Regression
Description: Generalizes logistic regression to multi-class classification problems.
Use Case: When predicting an outcome with more than two categories.
Relationship: Uncovers relationships between the log-odds of each category of the dependent variable and independent variables.

5. Lasso Regression
Description: A type of linear regression that includes L1 regularization, which can shrink some coefficients to zero, thus performing variable selection.
Use Case: When you have many predictors and want to select a subset that has the most influence on the outcome.
Relationship: Uncovers linear relationships but also performs feature selection.

6. Poisson Regression
Description: Used for modeling count data and contingency tables; assumes the dependent variable follows a Poisson distribution.
Use Case: When predicting the number of times an event occurs.
Relationship: Uncovers relationships between the rate of occurrence of the dependent variable and independent variables.

7. Ridge Regression
Description: A type of linear regression that includes L2 regularization, which prevents overfitting by shrinking coefficients.
Use Case: When you have multicollinearity or need to prevent overfitting in multiple linear regression.
Relationship: Uncovers linear relationships with regularization to prevent overfitting.

8. Polynomial Regression
Description: Extends linear regression by adding polynomial terms to model non-linear relationships.
Use Case: When the relationship between dependent and independent variables is non-linear.
Relationship: Uncovers polynomial relationships by fitting a polynomial equation to the data.

9. Bayesian Models
Description: Incorporate prior beliefs and update these with data to form posterior distributions.
Use Case: When you have prior knowledge or need to quantify uncertainty in predictions.
Relationship: Can uncover various types of relationships (linear, non-linear) with the incorporation of prior information.

10. Fixed-effects Regression
Description: Models where the individual-specific effect is treated as fixed and correlated with the independent variables.
Use Case: When you have panel data and want to control for unobserved heterogeneity that varies between entities but not over time.
Relationship: Uncovers within-entity relationships, accounting for entity-specific effects.

11. Random-effects Regression
Description: Models where the individual-specific effect is treated as a random variable uncorrelated with the independent variables.
Use Case: When you have panel data and want to generalize inferences beyond the sample.
Relationship: Uncovers both within-entity and between-entity relationships, assuming entity-specific effects are random.

12. Mixed-effects Regression
Description: Combines fixed and random effects; suitable for hierarchical or nested data structures.
Use Case: When you have data with both fixed and random effects (e.g., repeated measures).
Relationship: Uncovers relationships that account for both fixed effects and random variations.

13. Gaussian Process Regression
Description: A non-parametric, Bayesian approach to regression that models data using a Gaussian process.
Use Case: When modeling complex, non-linear relationships with uncertainty quantification.
Relationship: Uncovers flexible, non-linear relationships with predictions given as distributions.
